```python
#Завдання 1. Створіть одновимірний масив (вектор) зі зростаючою геометричною прогресією, де 
#кожне наступне число більше попереднього в 2 рази, починаючи з 2, та виведіть його 
#значення
import numpy as np

start = 2
num_elements = 10  # Кількість елементів у масиві
ratio = 2

geometric_progression = start * (ratio ** np.arange(num_elements))

print("Одновимірний масив зі зростаючою геометричною прогресією:")
print(geometric_progression)
```

```python
#Завдання 2.Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його оберненими 
#значеннями натуральних чисел, починаючи з 1, та виведіть його значення
import numpy as np

natural_numbers = np.arange(1, 10)

reciprocals = 1 / natural_numbers

matrix = reciprocals.reshape(3, 3)

print("Матриця 3x3, заповнена оберненими значеннями натуральних чисел:")
print(matrix)
```

```python
#Завдання 3.Створіть масив розміром 3x3, заповніть його випадковими дійсними числами в 
#діапазоні від -3 до 3 та знайдіть максимальний елемент
import numpy as np
matrix = np.random.uniform(-3, 3, (3, 3))

max_element = np.max(matrix)

print("Масив 3x3, заповнений випадковими числами в діапазоні від -3 до 3:")
print(matrix)
print("Максимальний елемент у масиві:")
print(max_element)
```

```python
#Завдання 4. Створіть масив розміром 3x3, заповніть його випадковими дійсними числами в 
#діапазоні від 0 до 1 та знайдіть вектор-стовпчик, що містить суму елементів кожного 
#стовпця, менших за 0.5
import numpy as np

matrix = np.random.rand(3, 3)

filtered_matrix = np.where(matrix < 0.5, matrix, 0)

column_sums = np.sum(filtered_matrix, axis=0)

print("Масив 3x3, заповнений випадковими числами в діапазоні від 0 до 1:")
print(matrix)
print("Вектор-стовпчик із сумами елементів у кожному стовпці, менших за 0.5:")
print(column_sums)
```

```python
#Завдання 5. Створіть два одновимірних масиви розміром 6, заповніть їх випадковими дійсними 
#числами від -2 до 2 та виконайте на них поелементні операції додавання, віднімання 
#та множення
import numpy as np

array1 = np.random.uniform(-2, 2, 6)
array2 = np.random.uniform(-2, 2, 6)

addition = array1 + array2

subtraction = array1 - array2

multiplication = array1 * array2

print("Перший масив:")
print(array1)
print("Другий масив:")
print(array2)
print("Результат поелементного додавання:")
print(addition)
print("Результат поелементного віднімання:")
print(subtraction)
print("Результат поелементного множення:")
print(multiplication)
```

```python
#Завдання 8. Створіть матрицю розміром 2x2, заповніть її випадковими цілими числами в діапазоні 
#від 1 до 15 та знайдіть її обернену матрицю.
import numpy as np

matrix = np.random.randint(1, 16, (2, 2))

inverse_matrix = np.linalg.inv(matrix)

print("Початкова матриця 2x2:")
print(matrix)
print("Обернена матриця:")
print(inverse_matrix)
```

```python
#Завдання 14. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в 
#діапазоні від -5 до 5 та знайдіть суму елементів матриці, які більше за їх середнє 
#арифметичне значення.
import numpy as np

matrix = np.random.uniform(-5, 5, (3, 3))

mean_value = np.mean(matrix)

greater_than_mean = matrix[matrix > mean_value]

sum_greater_than_mean = np.sum(greater_than_mean)

print("Матриця 3x3:")
print(matrix)
print("Середнє арифметичне значення матриці:")
print(mean_value)
print("Сума елементів, які більші за середнє арифметичне:")
print(sum_greater_than_mean)
```

```python
#Завдання 15. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в 
#діапазоні від 0 до 1 та знайдіть вектор-стовпчик, що містить найменше ціле значення 
#в кожному рядку матриці.
import numpy as np

matrix = np.random.rand(3, 3)

min_integers = np.floor(matrix).astype(int)  # Перетворюємо на ціле число, округлюючи вниз
min_values_per_row = np.min(min_integers, axis=1)  # Знаходимо мінімальне значення в кожному рядку

min_values_column = min_values_per_row.reshape(-1, 1)

print("Матриця 3x3:")
print(matrix)
print("Вектор-стовпчик, що містить найменше ціле значення в кожному рядку:")
print(min_values_column)
```
